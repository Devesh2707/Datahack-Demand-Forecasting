```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import keras

plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('new_df.csv')
```

```python
df['DateType'] = pd.to_datetime(df['DateType'])
```

```python
df['Day'] = [date.day for date in df['DateType']]
```

```python
df['Year'] = [date.year for date in df['DateType']]
```

```python
label_encode = ['store_id', 'sku_id', 'Month', 'Day', 'Year']
```

```python
selector = [
    'store_id',
    'sku_id',
    'total_price',
    'base_price',
    'is_featured_sku',
    'is_display_sku',
    'featured_display',
    'Month',
    'Median Sales',
    'LaggedSales',
    'LaggedAvailable',
    'Sales_dif',
#     'Day',
    'Year',
#     'store_product',
#     'total_price_bin_10',
#     'total_price_bin_5',
#     'base_price_bin_10',
#     'base_price_bin_5'
    'Split'
]
```

```python
df_selected = df[selector]
```

```python
for label in label_encode:
    df_selected = pd.concat([df_selected, pd.get_dummies(df[label])], axis = 1)
```

```python
df_selected.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = ['store_id', 'sku_id', 'Month',  'Year']
encoder = LabelEncoder()


for label in label_encode:
    df_selected[label] = encoder.fit_transform(df_selected[label])
```

```python
df_selected.head()
```

```python
train = df_selected[df_selected['Split'] == 'Train']
test = df_selected[df_selected['Split'] == 'Test']
```

```python
train = train.drop(['Split'], axis = 1)
```
